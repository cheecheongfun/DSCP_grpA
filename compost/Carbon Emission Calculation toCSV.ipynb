```python
import re

# Function to process tank values
def process_tank_value(value):
    if pd.isna(value):
        return value
    
    # Find all numbers (integer or decimal) before 'g' and divide them by 1000
    def replace_number(match):
        number = float(match.group(1))
        return f"{number / 1000}g"
    
    # Update regex to capture decimal numbers
    processed_value = re.sub(r'(\d+(\.\d+)?)g', replace_number, str(value))
    return processed_value

# Apply the function to all tank columns
tank_columns = ['Tank 1', 'Tank 2', 'Tank 3', 'Tank 4']
for col in tank_columns:
    df[col] = df[col].apply(process_tank_value)

# Display the updated DataFrame to verify changes
df.head()
```

```python
# Function to split, filter, and format each entry
def process_entry(entry):
    if isinstance(entry, str):
        items = [item.strip() for item in entry.split(',')]
        filtered_items = [item for item in items if 'spoon' not in item]
        return ''.join([f'[{item}]' for item in filtered_items])
    else:
        return np.nan

# Apply function to each tank column
df['Tank 1'] = df['Tank 1'].apply(process_entry)
df['Tank 2'] = df['Tank 2'].apply(process_entry)
df['Tank 3'] = df['Tank 3'].apply(process_entry)
df['Tank 4'] = df['Tank 4'].apply(process_entry)
#df['Tank 5'] = df['Tank 5'].apply(process_entry)

# Create a new DataFrame with Date and processed tank data
new_df = pd.DataFrame({
    'Date': df['Date'],
    'Tank 1': df['Tank 1'],
    'Tank 2': df['Tank 2'],
    'Tank 3': df['Tank 3'],
    'Tank 4': df['Tank 4'],
    #'Tank 5': df['Tank 5']
})
new_df
```

### Checking Invalid Food Types that are not Identified

```python
import re
import numpy as np
import pandas as pd

# Example DataFrames
# ref = pd.DataFrame({
#     'Food Name': ['eggshell', 'banana', 'beef'],
#     'Emissions (kg CO2/kg)': [0.01, 0.4, 27.0]
# })

# new_df = pd.DataFrame({
#     'Tank 1': ['100g of crushed eggshell', '50g of banana'],
#     'Tank 2': ['200g of beef', '150g of eggshell'],
#     'Tank 3': ['30g of banana', '20g of egg shells']
# })

# Convert emission data to a dictionary for easy lookup
emissions_dict = ref.set_index('Material')['CompostedD'].to_dict()

# Function to calculate the carbon offset for a given entry
def calculate_carbon_offset(entry):
    if pd.isna(entry):
        return np.nan
    # Extract food items and amounts
    items = re.findall(r'\[(.*?)\]', entry.lower())
    total_offset = 0
    for item in items:
        # Extract amount and food name
        match = re.match(r'(\d+\.?\d*)g (.*)', item)
        if match:
            amount = float(match.group(1))  # Amount in grams
            food_name = match.group(2).strip()  # Normalize food name
            # Check if any key from emissions_dict is in food_name
            matched = False
            for key in emissions_dict:
                if key in food_name:
                    offset = emissions_dict[key] * (amount)  # Convert grams to kg
                    total_offset += offset
                    matched = True
                    break
            if not matched:
                print(f"No match found for: {food_name}")
    return total_offset

# Apply the function to each tank column and date
for tank in ['Tank 1', 'Tank 2', 'Tank 3','Tank 4']:
    new_df[f'{tank} - C02'] = new_df[tank].apply(calculate_carbon_offset)

new_df
```
