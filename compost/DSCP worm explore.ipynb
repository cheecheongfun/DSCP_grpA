```python
def diagnostic_plots(df, variable):
    
    plt.figure(figsize=(15,6))
    plt.subplot(1, 3, 1)
    plt.rcParams["patch.edgecolor"] = "none"
    sns.histplot(df[variable], bins=30, kde = True, stat ="density", kde_kws=dict(cut = 3),alpha = 0.30)
    plt.title('Histogram')

    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])

    plt.show()
    
    # Count outliers using IQR method
    Q1 = df[variable].quantile(0.25)
    Q3 = df[variable].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[variable] < lower_bound) | (df[variable] > upper_bound)]

    print(f'Number of outliers in {variable}: {len(outliers)}')
    
diagnostic_plots(macph01, 'Temp01')
```

### Relationship between variables

```python
# Extract date and store it in a new column
macph01['date'] = macph01['Timestamp'].dt.date
macph01['time'] = macph01['Timestamp'].dt.time

macph01.head()
```

```python
macph01.tail()
```

```python
unique_dates = macph01['date'].unique()
unique_dates
```

```python
macph01['hour'] =macph01['Timestamp'].dt.strftime('%H:%M:%S')

# Define times to plot
times = ['10:00:00', '11:00:00', '12:00:00', '13:00:00', '14:00:00', '15:00:00',
         '16:00:00', '17:00:00', '18:00:00', '19:00:00', '20:00:00', '21:00:00']

# Plotting
fig, axs = plt.subplots(3, 4, figsize=(30, 10))

for idx, time in enumerate(times):
    row = idx // 4
    col = idx % 4
    filtered_df = macph01[macph01['hour'] == time]
    axs[row, col].plot(filtered_df['date'].apply(lambda x: x.strftime('%Y-%m-%d')), filtered_df['Temp01'], marker='o')
    axs[row, col].set_title(time)
    axs[row, col].set_xlabel('Date')
    axs[row, col].set_ylabel('Temperature')

plt.tight_layout()
plt.show()
```

```python
#Understanding the correlation of variables by showing the correlation matrix on a heatmap 
macph01_num = macph01.select_dtypes(['float64']).copy()

# Compute the correlation matrix
correlation_matrix = macph01_num.corr()

# Set up the matplotlib figure
plt.figure(figsize=(16, 16))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)

# Show the plot
plt.title('Correlation Heatmap of Macpherson Rack 01 Numerical Features')
plt.show()
```
