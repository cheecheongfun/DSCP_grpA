```python
# Convert the "hourly_interval" column to datetime format
tank1_data['hourly_interval'] = pd.to_datetime(tank1_data['hourly_interval'])
tank2_data['hourly_interval'] = pd.to_datetime(tank2_data['hourly_interval'])
tank3_data['hourly_interval'] = pd.to_datetime(tank3_data['hourly_interval'])
tank4_data['hourly_interval'] = pd.to_datetime(tank4_data['hourly_interval'])

# Initialize the "feed" column in each tank dataset with empty strings
tank1_data['feed'] = ""
tank2_data['feed'] = ""
tank3_data['feed'] = ""
tank4_data['feed'] = ""

# Update the "feed" column in each tank dataset based on the feeding log
for index, row in feeding_log.iterrows():
    date = row['Date']
    tank1_feed = row['Tank 1']
    tank2_feed = row['Tank 2']
    tank3_feed = row['Tank 3']
    tank4_feed = row['Tank 4']
    
    tank1_data.loc[tank1_data['hourly_interval'].dt.date == date.date(), 'feed'] = tank1_feed
    tank2_data.loc[tank2_data['hourly_interval'].dt.date == date.date(), 'feed'] = tank2_feed
    tank3_data.loc[tank3_data['hourly_interval'].dt.date == date.date(), 'feed'] = tank3_feed
    tank4_data.loc[tank4_data['hourly_interval'].dt.date == date.date(), 'feed'] = tank4_feed
```

```python
# Export the updated datasets to CSV files
tank1_data.to_csv('Updated_Tank1_Data.csv', index=False)
tank2_data.to_csv('Updated_Tank2_Data.csv', index=False)
tank3_data.to_csv('Updated_Tank3_Data.csv', index=False)
tank4_data.to_csv('Updated_Tank4_Data.csv', index=False)

# Display the head of each dataset to verify
tank1_data.head(), tank2_data.head(), tank3_data.head(), tank4_data.head()
```

```python
# Combine all tank data into a single DataFrame
combined_tank_data = pd.concat([tank1_data, tank2_data, tank3_data, tank4_data], ignore_index=True)

# Save the combined data to a new CSV file
combined_tank_data.to_csv('Combined_Tank_Data.csv', index=False)


# Show the head of the combined data
combined_tank_data.head()
```

```python
file_path_feeding_logs = 'MacPherson Feeding Logs.xlsx'
excel_file_feeding_logs = pd.ExcelFile(file_path_feeding_logs)

# List all sheet names in the new file
sheet_names_feeding_logs = excel_file_feeding_logs.sheet_names
sheet_names_feeding_logs
```

```python
feeding_log_data = pd.DataFrame()

# Sheets 2 to 4 are 'Sep to Dec 2023', 'Jan to Jun 2024', and 'Jul to Dec 2024'
sheets_to_combine = ['Sep to Dec 2023', 'Jan to Jun 2024', 'Jul to Dec 2024']

for sheet in sheets_to_combine:
    data = pd.read_excel(file_path_feeding_logs, sheet_name=sheet)
    data['Source Sheet'] = sheet  # Add a column for the sheet name
    feeding_log_data = pd.concat([feeding_log_data, data], ignore_index=True)

# Save the combined dataframe to a new Excel file
output_feeding_log_file_path = 'Feeding_Log_Combined.xlsx'
feeding_log_data.to_excel(output_feeding_log_file_path, index=False)

feeding_log_data.head()
```

```python
rack_data_path = 'Combined_MacPherson_Rack_1.xlsx'
feeding_log_path = 'Feeding_Log_Combined.xlsx'

# Load the Excel files
rack_data = pd.read_excel(rack_data_path)
feeding_log_data = pd.read_excel(feeding_log_path)

# Convert the 'Date' columns to strings and strip any leading or trailing spaces
rack_data['Date'] = rack_data['Date'].astype(str).str.strip()
feeding_log_data['Date'] = feeding_log_data['Date'].astype(str).str.strip()

# Ensure that the Date columns are in datetime format
rack_data['Date'] = pd.to_datetime(rack_data['Date'], format='%d %B %Y', errors='coerce')
feeding_log_data['Date'] = pd.to_datetime(feeding_log_data['Date'], format='%d %B %Y', errors='coerce')

# Filter the feeding log data for Tank 1 and include the correct pH column
feeding_log_tank1_ph = feeding_log_data[['Date', 'Tank 1', 'pH values of worm feed']].rename(columns={'Tank 1': 'Feed', 'pH values of worm feed': 'pH of Worm Feed'})

# Combine the feeding log data for Tank 1 with pH with the rack data based on the Date
combined_data_tank1_ph = pd.merge(rack_data, feeding_log_tank1_ph, on='Date', how='left')

# Save the combined dataframe to a new Excel file
output_combined_tank1_ph_path = 'Combined_MacPherson_Rack_1_with_Feed_and_pH.xlsx'
combined_data_tank1_ph.to_excel(output_combined_tank1_ph_path, index=False)


combined_data_tank1_ph.head()
```

```python
combined_all_tanks_with_feed_path = 'Combined_All_Tanks_with_Feed.xlsx'
combined_all_tanks_with_feed = pd.read_excel(combined_all_tanks_with_feed_path)

# Function to aggregate columns
def aggregate_columns(df, prefix):
    col1 = f"{prefix}01"
    col2 = f"{prefix}02"
    new_col = f"Soil - {prefix.capitalize()}"
    if col1 in df.columns and col2 in df.columns:
        df[col1] = pd.to_numeric(df[col1], errors='coerce')
        df[col2] = pd.to_numeric(df[col2], errors='coerce')
        df[new_col] = (df[col1] + df[col2]) /2
        df.drop(columns=[col1, col2], inplace=True)

# Function to aggregate pH columns
def aggregate_ph_columns(df):
    col1 = 'pH Rod 1'
    col2 = 'pH Rod 2'
    new_col = 'Soil - PH'
    if col1 in df.columns and col2 in df.columns:
        df[col1] = pd.to_numeric(df[col1], errors='coerce')
        df[col2] = pd.to_numeric(df[col2], errors='coerce')
        df[new_col] = (df[col1] + df[col2]) /2
        df.drop(columns=[col1, col2], inplace=True)

# List of prefixes to aggregate
prefixes = ['Nitrogen', 'Phosphorous', 'Potassium', 'SoilM']

# Aggregate columns
for prefix in prefixes:
    aggregate_columns(combined_all_tanks_with_feed, prefix)

# Aggregate pH columns
aggregate_ph_columns(combined_all_tanks_with_feed)

# Drop heat and humidity columns if they exist
columns_to_drop = ['Heat01', 'Hum01', 'Number of Worms (non-counted)', 'Light Intensity', 'Grams of Compost Harvested', 'Compost Notes', 'Buzzer','Date']
columns_to_drop = [col for col in columns_to_drop if col in combined_all_tanks_with_feed.columns]
combined_all_tanks_with_feed.drop(columns=columns_to_drop, inplace=True)

combined_all_tanks_with_feed.rename(columns={'Soil - Soilm': 'Soil - Moisture', 'Feed':'feed','Timestamp':'hourly_interval','Soil - Phosphorous':'Soil - Phosphorus','Temp01':'Soil - Temperature'}, inplace=True)

# Display the updated dataframe
print(combined_all_tanks_with_feed.head())

# Save the aggregated dataframe to a new Excel file
output_path = 'Aggregated_All_Tanks_with_Feed.xlsx'
combined_all_tanks_with_feed.to_excel(output_path, index=False)
```
